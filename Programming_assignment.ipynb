```python
# Code for 2.3
distances = [5.657, 4.243, 2.828, 2.236, 5, 1.414, 3.606, 3.606, 4.123, 1.414]
# DO NOT CHANGE ANYTHING ABOVE. FILL IN YOUR CODE BELOW.
def class_label_predictor(distance, class_label, length, n):
  if n>length: #Checking whether the neighbour value is greater than the length of the data or not.
    return class_label_predictor(distance, class_label, length, 10) #If it is greater than it will be put to the maximum value ie, length
  else:
    res = {distances[i]: class_labels[i] for i in range(len(distances))} #combined distances and their corresponding class.
    sorted_res = dict(sorted(res.items())) # Sorted the dictinoray according to the key
    first_n_terms = {k: sorted_res[k] for k in list(sorted_res)[:n]} #Extracting the first n neighbours with the smallest values for distance.
    n_list_class = list(first_n_terms.values()) # Taking the first n neighbours class values and converting it into a list.
    class_freq = {p:n_list_class.count(p) for p in n_list_class} # To find the class freq occurance of the first selected n neighbours.

    return max(class_freq, key=class_freq.get) # Returning out the key with the maximum class freq within in the neighbours


euc_distance = xy_data['distances'].to_list()
classes = xy_data['class_labels'].to_list()
length = xy_data.shape[0]

class_label = class_label_predictor(euc_distance, classes, length, 3) # Output of the function is stored in the class label.

# DO NOT CHANGE ANYTHING BELOW. FILL IN YOUR CODE ABOVE.
print("class label : ", class_label)
```

```python
# Code for 3.1
def split_df_values(df, df_column, target):
  a = []
  Children_list = list(df[df_column].value_counts().keys())
  for i in Children_list:
    a.append(df[df[df_column] == i][target].value_counts().to_list())
  return a


def split_data(df, df_column):
  a = []
  Children_list = list(data[df_column].value_counts().keys())
  for i in Children_list:
    a.append(df[df[df_column] == i])
  return a


def gini_index_split(a):
  if len(a) == 1:
    return 0
  else:
    return 1-(a[0]/sum(a))**2 -(a[1]/sum(a))**2


def gini_impurity(df, data_column, target): 
  column_values = df[data_column].value_counts()
  gini_total = 0 
  for key in column_values.keys():
      data_value = data[target][data[data_column] == key].value_counts()
      n = column_values[key]
      shape = data.shape[0]
      gini_total = gini_total + (( n / shape) * gini_index_split(list(data_value)))
  print("Gini Impurity =",gini_total)
  return gini_total


def entropy(df, df_column):
  total_values = len(df[df_column])
  column_values = df[df_column].value_counts()
  h_key = 0
  for key in column_values.keys():
    prob_key = column_values[key]/total_values
    log_value = math.log(prob_key, 2)
    h_key = h_key + (prob_key * log_value)
  print("Entropy = ", df_column, h_key*(-1))
  return h_key*(-1)


def info_gain(data, target): # Parent
  info_gain = []

  class_label_values = data[target].value_counts()
  IG_Class_label = 0
  for class_label in class_label_values.keys():
    prob_key = class_label_values[class_label]/len(data)
    IG_Class_label += (prob_key * math.log(prob_key, 2))
  return -IG_Class_label


def information_gain(data, df_column, target): # Function for finding it 
    column_values = data[df_column].value_counts()
    entropy_key = {}
    inf_sum = 0
    for column_value in column_values.keys():
      decision_values = data[target][data[df_column] == column_value].value_counts()
      h_key = 0
      total_count = 0
      for keys in decision_values.keys():
        total_count = total_count + decision_values[keys]
      for key in decision_values.keys():
        calc_value = decision_values[key]/total_count
        h_key += calc_value * math.log(calc_value,2)
        entropy_key[column_value] = -h_key
      total_count = len(data)
      probability_column_value = (column_values[column_value]/total_count)
      inf_sum += (probability_column_value) * (entropy_key[column_value])
    return inf_sum


import math
def entropy(df, df_column):
  total_values = len(df[df_column])
  column_values = df[df_column].value_counts()
  h_key = 0
  for key in column_values.keys():
    prob_key = column_values[key]/total_values
    log_value = math.log(prob_key, 2)
    h_key = h_key + (prob_key * log_value)
  print("Entropy = ", df_column, h_key*(-1))
  return h_key*(-1)


gini = []
Entropy = []
Informationgain = []
for i in list(data.columns[:-1]):
  [display(i) for i in split_data(data, i)]
  print("Decision Making =", split_df_values(data, i, "Play Golf"))
  [print("Gini Index = ",gini_index_split(p)) for p in split_df_values(data, i, "Play Golf")]
  gini.append(gini_impurity(data, i, "Play Golf"))
  Entropy.append(entropy(data,i))
  print("Information Gain = ", info_gain(data, "Play Golf")-information_gain(data, i, "Play Golf"))
  Informationgain.append(info_gain(data, "Play Golf")-information_gain(data, i, "Play Golf"))
print("Gini Impurity of all columns =", gini)
print("Entropy of all columns =",Entropy)
print("Information Gain of all columns = ", Informationgain)
```
